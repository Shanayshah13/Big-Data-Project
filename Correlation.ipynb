```python
import pandas as pd
from collections import defaultdict
```

```python
inputFile = "filterd_data.tsv"
data = pd.read_csv(inputFile,sep='\t',low_memory=False)
```

```python
data
```

```python
data_correlated = data.corr()
```

```python
data_correlated
```

```python
data_correlated.columns
```

```python
hashtableCorrelation = defaultdict(int)
columns = data_correlated.columns
for i in range(len(data_correlated)):
    row = data_correlated.iloc[i,:]
    #print(row)
    minCol, maxCol = '',''
    maxVal = float('-inf')
    minVal = float('inf')
    for j in range(len(row)):
        #print(row[j])
        if row[j]==1:
            continue
        if row[j]>maxVal: 
            maxVal = j
            maxCol = columns[j]
        if row[j]<minVal: 
            minVal = j
            minCol = columns[j] 
    hashtableCorrelation[maxCol]+=1
    hashtableCorrelation[minCol]+=1
```

```python
print(len(hashtableCorrelation))
```

```python
sorted_hashtable = sorted(hashtableCorrelation.items(), key=lambda kv: kv[1],reverse=True)
sorted_hashtable
```
